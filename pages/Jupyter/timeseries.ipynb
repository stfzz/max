```python
cont_c = 0
cont_a = 0
datafineprecedenti_c = []
datafineprecedenti_a = []

for ind in dfx.index:
    
    if dfx['utente'][ind] == "comunale":
        cont_c += 1
        datafineprecedenti_c.append(str(dfx["DataFine"][ind]))
        d_c = copy.copy(datafineprecedenti_c)
        for a in datafineprecedenti_c:
            if str(dfx["DataInizio"][ind]) > a:
                print()
                print(dfx["Cognome"][ind] + " " + dfx["Nome"][ind] +" cont_c = " + str(cont_c))
                cont_c -= 1
                print("Sottrazione cont_c = " + str(cont_c))
                d_c.remove(a)
        datafineprecedenti_c = copy.copy(d_c)
        
        
    elif dfx['utente'][ind] == "aziendale":
        cont_a += 1
        datafineprecedenti_a.append(str(dfx["DataFine"][ind]))
        d_a = copy.copy(datafineprecedenti_a)
        for a in datafineprecedenti_a:
            if str(dfx["DataInizio"][ind]) > a:
                print()
                print(dfx["Cognome"][ind] + " " + dfx["Nome"][ind] +" cont_a = " + str(cont_a))
                cont_a -= 1
                print("Sottrazione cont_a = " + str(cont_a))
                d_a.remove(a)
                print()
        datafineprecedenti_a = copy.copy(d_a)
        
        
    print("bimbo " + dfx["utente"][ind] + "\t--> " + dfx["Cognome"][ind] + " " + dfx["Nome"][ind] + " \t--> cont_c = " + str(cont_c) + " and cont_a = " + str(cont_a))    
    
    if (cont_c >= cont_a) and dfx["utente"][ind] == "comunale":
        print()
        print("Bimbo  " + dfx["Cognome"][ind] + " " + dfx["Nome"][ind] + "  \tin eccesso COMUNALE, in data " + str(dfx["DataInizio"][ind]))
        cont_c -= 1
        print("tolto bimbo: cont_c = " + str(cont_c) + " and cont_a = " + str(cont_a)) 
        print()
print()    
print("################### END OF ITERATION ###################")
    
    
```
